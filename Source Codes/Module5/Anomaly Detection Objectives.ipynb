```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulate network traffic data
normal_traffic = np.random.normal(100, 10, 100)
anomalous_traffic = np.random.normal(130, 10, 10)

# Create a histogram to visualize traffic distribution
plt.hist(normal_traffic, bins=20, alpha=0.7, label='Normal Traffic')
plt.hist(anomalous_traffic, bins=20, alpha=0.7, label='Anomalous Traffic')
plt.xlabel('Traffic Volume')
plt.ylabel('Frequency')
plt.legend()
plt.title('Distribution of Normal and Anomalous Traffic')
plt.show()
```

```python
# Simulate easy configuration by allowing users to set a threshold interactively
threshold = int(input("Enter traffic threshold: "))

# Visualize the traffic data with the chosen threshold
plt.hist(normal_traffic, bins=20, alpha=0.7, label='Normal Traffic')
plt.axvline(x=threshold, color='red', linestyle='--', label='Threshold')
plt.xlabel('Traffic Volume')
plt.ylabel('Frequency')
plt.legend()
plt.title('Configurable Threshold for Anomaly Detection')
plt.show()
```

```python
# Simulate changing data trends with a daily pattern
timestamps = pd.date_range(start='2023-01-01', periods=100, freq='H')
traffic = np.sin(np.linspace(0, 2 * np.pi, 100)) * 100 + 200

# Visualize the traffic data with changing trends
plt.plot(timestamps, traffic)
plt.xlabel('Timestamp')
plt.ylabel('Traffic Volume')
plt.title('Adapting to Changing Data Trends (e.g., Daily Pattern)')
plt.show()
```

```python
# Simulate different types of data: web, email, and database traffic
service_types = ['Web', 'Email', 'Database']
traffic = [np.random.normal(100, 10, 100) for _ in service_types]

# Create box plots to compare traffic across different service types
plt.boxplot(traffic, labels=service_types)
plt.ylabel('Traffic Volume')
plt.title('Versatility Across Different Service Types')
plt.show()
```
