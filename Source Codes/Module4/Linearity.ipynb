```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Generate random data for 5 features and an outcome variable
np.random.seed(0)
n_samples = 100
feature1 = np.random.rand(n_samples)
feature2 = np.random.rand(n_samples)
feature3 = np.random.rand(n_samples)
feature4 = np.random.rand(n_samples)
feature5 = np.random.rand(n_samples)

# Generate labels
outcome = 2 * feature1 + 3 * feature2 + 1.5 * feature3 + 0.5 * feature4 + 0.2 * feature5 + np.random.randn(n_samples)

# Create scatterplots for each feature vs. the outcome with regression lines
plt.figure(figsize=(15, 3))

for i, feature in enumerate([feature1, feature2, feature3, feature4, feature5], start=1):
    plt.subplot(150 + i)
    plt.scatter(feature, outcome)
    plt.xlabel(f"Feature {i}")
    plt.ylabel("Outcome")
    
    # Fit a linear regression model
    slope, intercept, r_value, _, _ = linregress(feature, outcome)
    
    # Plot the regression line
    plt.plot(feature, intercept + slope * feature, color='red', linewidth=2)
    
    # Display R-squared value on the plot
    plt.text(0.7, 0.1, f'R-squared = {r_value**2:.2f}', transform=plt.gca().transAxes, fontsize=10, color='red')

plt.tight_layout()
plt.show()
```
